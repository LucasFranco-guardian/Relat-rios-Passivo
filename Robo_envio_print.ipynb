```python
import pygetwindow as gw
import pyautogui
import os
import time
import schedule
from datetime import datetime
import smtplib
from email.mime.multipart import MIMEMultipart
from email.mime.base import MIMEBase
from email.mime.text import MIMEText
from email import encoders
```

```python
# Caminho onde os prints ser√£o salvos
PASTA_DESTINO = r"C:\Users\LucasCavalcante\Documents\Prints_GARE11"
NOME_JANELA = 'Broadcast'

def garantir_pasta_existente(pasta):
    os.makedirs(pasta, exist_ok=True)

def encontrar_janela(nome_janela):
    for w in gw.getWindowsWithTitle(nome_janela):
        if nome_janela.upper() in w.title.upper():
            return w
    return None

def preparar_janela_para_print(janela):
    try:
        janela.minimize()
        time.sleep(1)
        janela.restore()
        time.sleep(2)  # tempo extra para garantir que a janela esteja vis√≠vel
    except Exception as e:
        print(f"‚ö†Ô∏è Erro ao restaurar a janela: {e}")

def tirar_screenshot_da_janela(janela, pasta_destino, nome_arquivo='print_PARCIAL_GARE11.png'):
    try:
        if janela.left == 0 and janela.top == 0 and janela.width == 0 and janela.height == 0:
            raise Exception("Coordenadas inv√°lidas da janela")

        x, y = janela.left, janela.top
        w_, h_ = janela.width, janela.height

        caminho_arquivo = os.path.join(pasta_destino, nome_arquivo)
        screenshot = pyautogui.screenshot(region=(x, y, w_, h_))
        screenshot.save(caminho_arquivo)

        print(f"‚úÖ Screenshot salvo em: {caminho_arquivo}")
        return caminho_arquivo

    except Exception as e:
        print(f"‚ùå Erro ao tirar o print: {e}")
        return None

# Fun√ß√£o para enviar e-mail com a imagem anexada
def send_email_with_image(image_path):
    sender_email = 'lc@guardian-asset.com'
    sender_password = 'wwafykwsxuhslwwv'
    recipient_email = ['lc@guardian-asset.com']

    msg = MIMEMultipart()
    msg['From'] = sender_email
    msg['To'] = ", ".join(recipient_email)
    msg['Subject'] = "PRINT PARCIAL - CORRETORAS - GARE11"
    msg.attach(MIMEText("SEGUE: PRINT PARCIAL - CORRETORAS - GARE11", 'plain'))

    try:
        with open(image_path, "rb") as attachment:
            part = MIMEBase("image", "png")
            part.set_payload(attachment.read())
        encoders.encode_base64(part)
        part.add_header("Content-Disposition", f"attachment; filename= {os.path.basename(image_path)}")
        msg.attach(part)
    except Exception as e:
        print(f"‚ùå Falha ao anexar o arquivo: {e}")
        return

    try:
        with smtplib.SMTP_SSL('smtp.gmail.com', 465) as smtp_server:
            smtp_server.login(sender_email, sender_password)
            smtp_server.sendmail(sender_email, recipient_email, msg.as_string())
        print(f"üìß E-mail enviado para: {', '.join(recipient_email)}")
    except Exception as e:
        print(f"‚ùå Falha ao enviar e-mail: {e}")


def tarefa_automatica():
    print("üïê Iniciando rotina de captura + envio de print...")
    janela = encontrar_janela(NOME_JANELA)
    if janela:
        preparar_janela_para_print(janela)
        caminho_print = tirar_screenshot_da_janela(janela, PASTA_DESTINO)
        if caminho_print:
            send_email_with_image(caminho_print)
    else:
        print(f"‚ùå Janela '{NOME_JANELA}' n√£o encontrada.")

schedule.every(30).minutes.do(tarefa_automatica)

tarefa_automatica()

while True:
    schedule.run_pending()
    time.sleep(5)
```
