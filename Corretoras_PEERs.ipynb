```python
df_compras = df.drop(columns=['Vendas'])
df_vendas = df.drop(columns=['Compras'])
```

```python
df_vendas_gare11 = (df_vendas.query('Fundo == "GARE11"').sort_values(['Vendas'], ascending=False).iloc[:7, :])
df_vendas_trxf11 = (df_vendas.query('Fundo == "TRXF11"').sort_values(['Vendas'], ascending=False).iloc[:7, :])
df_vendas_hgru11 = (df_vendas.query('Fundo == "HGRU11"').sort_values(['Vendas'], ascending=False).iloc[:7, :])
df_vendas_knri11 = (df_vendas.query('Fundo == "KNRI11"').sort_values(['Vendas'], ascending=False).iloc[:7, :])
df_vendas_alzr11 = (df_vendas.query('Fundo == "ALZR11"').sort_values(['Vendas'], ascending=False).iloc[:7, :])
df_vendas_btlg11 = (df_vendas.query('Fundo == "BTLG11"').sort_values(['Vendas'], ascending=False).iloc[:7, :])
df_vendas_hglg11 = (df_vendas.query('Fundo == "HGLG11"').sort_values(['Vendas'], ascending=False).iloc[:7, :])

df_compras_gare11 = (df_compras.query('Fundo == "GARE11"').sort_values(['Compras'], ascending=False).iloc[:7, :])
df_compras_trxf11 = (df_compras.query('Fundo == "TRXF11"').sort_values(['Compras'], ascending=False).iloc[:7, :])
df_compras_hgru11 = (df_compras.query('Fundo == "HGRU11"').sort_values(['Compras'], ascending=False).iloc[:7, :])
df_compras_knri11 = (df_compras.query('Fundo == "KNRI11"').sort_values(['Compras'], ascending=False).iloc[:7, :])
df_compras_alzr11 = (df_compras.query('Fundo == "ALZR11"').sort_values(['Compras'], ascending=False).iloc[:7, :])
df_compras_btlg11 = (df_compras.query('Fundo == "BTLG11"').sort_values(['Compras'], ascending=False).iloc[:7, :])
df_compras_hglg11 = (df_compras.query('Fundo == "HGLG11"').sort_values(['Compras'], ascending=False).iloc[:7, :])
```

```python
df_corretoras_V = df_vendas_gare11.drop(columns=['Fundo', 'Vendas'])
df_corretoras_V

fundos = [df_vendas_gare11, df_vendas_trxf11, df_vendas_hgru11, df_vendas_knri11, df_vendas_alzr11, df_vendas_btlg11, df_vendas_hglg11]
nomes_fundos = [df['Fundo'].iloc[0] for df in fundos]
corretoras_interesse = df_corretoras_V['Corretora'].tolist()

result = pd.DataFrame({'Corretoras': corretoras_interesse})
result.set_index('Corretoras', inplace=True)

for nome, df in zip(nomes_fundos, fundos):
    filtro = df[df['Corretora'].isin(corretoras_interesse)]
    tabela = filtro[['Corretora', 'Vendas']].groupby('Corretora').sum()
    result[nome] = tabela['Vendas']

result = result.fillna(0)

result = result.reset_index()

final_V = pd.DataFrame(result)

final_V
```
